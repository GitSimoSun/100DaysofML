```python
i = np.array([1, 0])
j = np.array([0, 1])
```

```python
rad_angle = np.arccos(i.dot(j) / (la.norm(i) * la.norm(j)))
rad_angle
```

```python
deg_angle = rad_angle * 180 / np.pi
deg_angle
```

```python
plt.plot(-0.2, -0.2, "None")

plt.plot(1.2, 1.2, "None")

plt.axis('equal')

plt.quiver(0, 0, *i, color=['r'], scale_units = 'xy', scale = 1)
plt.quiver(0, 0, *j, color=['b'], scale_units = 'xy', scale = 1)

plt.show()
```

```python
u = np.array([1.5, 1])
v = 2*i


projected_u = (u.dot(v) / la.norm(v) ** 2) * v


plt.plot(-0.2, -0.2, "None")

plt.plot(2, 0.7, "None")

plt.axis('equal')

plt.quiver(0, 0, *u, color=['r'], scale_units = 'xy', scale = 1)
plt.quiver(0, 0, *v, color=['b'], scale_units = 'xy', scale = 1)
plt.quiver(0, 0, *projected_u, color=['g'], scale_units = 'xy', scale = 1)
plt.vlines(u[0], v[1], u[1], linestyles="--", colors=["y"])


plt.show()
```
