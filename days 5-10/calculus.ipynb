```python
r = 1
x = np.linspace(-1, 1, 1000)
y = np.sqrt(r**2 - x**2)


plt.plot(x, y, c='b')
plt.plot(x, -y, c='b')

plt.axis('equal')


plt.title("Plotting a circle using x + y = 1")
plt.show()
```

```python
r = 1
theta = np.linspace(0, 2*np.pi, 100)

x = r * np.cos(theta)
y = r * np.sin(theta)


plt.plot(x, y, c='b')

plt.axis('equal')


plt.title("Plotting a circle using polar coordinates")
plt.show()
```

```python
sp.limit(sp.floor(x), x, 1, dir="-")
```

```python
discontinuity_points = []
for i in np.arange(0, 5.01, 0.01):
    right_lim = sp.limit(sp.floor(x), x, i, dir="+")
    left_lim = sp.limit(sp.floor(x), x, i, dir="-")
    if right_lim != left_lim:
        discontinuity_points.append(i)
        
print(discontinuity_points)
```

```python
x1 = np.linspace(0, 5, 1000)
y = np.floor(x1)

plt.plot(x1, y, label="floor")
plt.scatter(discontinuity_points, np.floor(discontinuity_points), c='orange', label="Discontinuity Points")

plt.legend()
plt.show()
```
